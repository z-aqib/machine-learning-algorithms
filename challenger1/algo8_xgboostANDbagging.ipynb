# Finding the best bagging
here we will run a for loop and continously find the best model with bagging

```python
best_model_roc = 0
best_model = None
best_bagging_count = 0
```

```python
def train_model(bagging_count):
    print("CASE", bagging_count)
    sample_model = BaggingClassifier(estimator=model, n_estimators=bagging_count, verbose=2)
    sample_model.fit(trainX, trainY)
    y_pred = sample_model.predict(testX)
    roc_score = roc_auc_score(testY, y_pred)
    accuracy = accuracy_score(testY, y_pred)
    print(" accuracy =", accuracy)
    return roc_score, sample_model
```

```python
bagging_count = 1
prev_roc_score = 0
while bagging_count <= 100:
    roc_score, sample_model = train_model(bagging_count)
    print(" roc =", roc_score)
    
    if roc_score > best_model_roc:
        best_model_roc = roc_score
        best_model = sample_model
        best_bagging_count = bagging_count
        print("selected as best")
    
    # If the ROC score is the same as the previous one, skip the next value
    if roc_score == prev_roc_score:
        bagging_count += 2  # Skip the next value by incrementing by 2
    else:
        bagging_count += 1  # Otherwise, increment by 1

    # Update prev_roc_score for the next iteration
    prev_roc_score = roc_score
```

```python
model = best_model
```
